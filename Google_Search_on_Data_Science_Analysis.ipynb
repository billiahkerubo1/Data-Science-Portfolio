# Importing Libraries

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
```

# Getting data

```python
# Google search trends on Data Science
trends.build_payload(kw_list=["Remote Jobs"])
data = trends.interest_by_region()
data = data.sort_values(by="Remote Jobs", ascending=False)
data = data.head(10)
print(data)
```

United States searches for remote jobs the most.

```python
# Visualizing the search
data.reset_index().plot(x="geoName", y="Remote Jobs", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
# Evaluating the trend change in the Google search for Data Science
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Remote Jobs'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data['Remote Jobs'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Remote Jobs', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

The search for remote jobs started increasing from the year 2020. This could be due to the COVID-19 pandemic.

The search for remote jobs was highest in 2022.
