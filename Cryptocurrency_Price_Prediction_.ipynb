# Importing Libraries

```python
# Importing libraries
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
from autots import AutoTS
```

# Getting Data

```python
# Getting Bitcoin prices data for the past 360 days
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=360)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('BTC-USD', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

# Understanding the data

```python
# Checking the size of the data
data.shape
```

There are 361 rows and 7 columns.

```python
# Visualizing the chang in Bitcoin prices
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Bitcoin Price Analysis", 
                     xaxis_rangeslider_visible=False)
figure.show()
```

```python
# Evaluating the correlation between the close column and other variables
correlation = data.corr()
print(correlation["Close"].sort_values(ascending=False))
```

```python
# Prediction
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

The predictions are based on historical data.
